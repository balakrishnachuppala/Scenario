```python
from faker import Faker 
import hashlib
f=Faker()
```

# Generating Dummy Data

```python
Faker.seed(111)

a=[]
s='' 
def fila(x):
    if x not in a:
        s=x
    else:
        s=fila(f.name())
    return s

for i in range(100000):
    n=f.name()
    a.append(fila(n) )

print('A Total  Count : {}, A Unique  Count : {} '.format(len(a),len(set(a))))
```

# Generation HASH and new deterministic data based on Hash

```python
 
b=[]
ss=[]

for j in range(len(a)): 
    s=hashlib.md5(a[j].encode()).hexdigest()
    ss.append(s)
    Faker.seed(s)  
    b.append(f.name())

d = {m:ss.count(m) for m in ss}
#print(d)
print('A Total  Count : {}, B Total  Count : {} '.format(len(a),len(b)))
print('A Unique Count : {}, B unique Count : {} '.format(len(set(a)),len(set(b))))
print('Hash total Count : {} , Hash Unique count : {} '.format(len(ss),len(set(ss))))
```

# Finding any dups in hash list

```python
print('Length of Hash list : {}, Lengh of unique hash : {}'.format(len(d),len(set(d))))
```

```python
def find_dups(dicts):
    my_dict = dicts
    Dictionary = dict()
    for (new_key, new_value) in my_dict.items():
        if new_value > 1:
            Dictionary[new_key] = new_value
    return Dictionary

print(find_dups(d))
```

# Finding dups in generated data based on hash

```python
bb = {m:b.count(m) for m in b}
print(find_dups(bb))
```

# Finding dups values for the data

```python
dstr='Robert Pena'
indx = []
for idx, val in enumerate(b):
    if dstr == val:
        indx.append(idx)         

print('Duplicate index for string {} : {}'.format(dstr,indx))
```
